# Task 242 implementation and verification

This notebook reads `task242.md`, summarizes the implementation requirements, and validates the `contracts/clinical-trial` contract changes by running the package tests.


```python
from pathlib import Path
import subprocess
import os

repo_root = Path('/workspaces/contracts')
print('Repository root:', repo_root)

# Attempt to install rustup if not already available.
if subprocess.run(['which', 'rustup'], capture_output=True, text=True).returncode != 0:
    print('Rustup not found; installing rustup...')
    install = subprocess.run(
        ['bash', '-lc', 'curl --proto "=https" --tlsv1.2 -sSf https://sh.rustup.rs | sh -s -- -y'],
        capture_output=True,
        text=True,
    )
    print('install return code:', install.returncode)
    print(install.stdout)
    print(install.stderr)
else:
    print('rustup already installed')

cargo_home = os.path.expanduser('~/.cargo/bin')
print('cargo_home:', cargo_home)
path = os.environ.get('PATH', '')
if cargo_home not in path:
    os.environ['PATH'] = cargo_home + os.pathsep + path
    print('PATH updated')

print('Which cargo now:')
print(subprocess.run(['which', 'cargo'], capture_output=True, text=True).stdout)
print('cargo --version:')
print(subprocess.run(['cargo', '--version'], capture_output=True, text=True).stdout)
print('rustc --version:')
print(subprocess.run(['rustc', '--version'], capture_output=True, text=True).stdout)

# Ensure latest stable toolchain is installed
print('Installing stable toolchain...')
subprocess.run(['bash', '-lc', 'export PATH="$HOME/.cargo/bin:$PATH" && rustup toolchain install stable -q'], shell=False)

# Run tests with rustup stable
print('\nRunning cargo test -p clinical-trial with rustup stable...')
result = subprocess.run(
    ['bash', '-lc', 'export PATH="$HOME/.cargo/bin:$PATH" && rustup default stable && cargo test -p clinical-trial'],
    cwd=str(repo_root),
    capture_output=True,
    text=True,
)
print('RETURN CODE:', result.returncode)
print(result.stdout)
print(result.stderr)
```
